```python
import pandas as pd
```

```python
metrics = pd.read_csv("mask_metrics.csv")
```

```python
metrics_pyradiomics = pd.read_csv("mask_metrics_pyradiomics.csv")
```

```python
metrics.sort_values(by="bdmap_id", inplace=True)
metrics_pyradiomics.sort_values(by="bdmap_id", inplace=True)
```

```python
for i, (_, row) in enumerate(metrics.iterrows()):
    other_row = metrics_pyradiomics.iloc[i]

    if not row.equals(other_row):
        print(f"\nUHOH at row {i}")

        print("\nmetrics:")
        print(row)

        print("\nmetrics_pyradiomics:")
        print(other_row)
```
